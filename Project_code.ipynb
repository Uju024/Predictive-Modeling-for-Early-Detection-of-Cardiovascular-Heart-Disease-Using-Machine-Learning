## Handling and Removing Outliers

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data = data[~((data < (Q1 - 1.5 * IQR))|(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

## Preprocessing

```python
# check the class balance after removing outliers

Target = data['target'] # target variable
class_counts = data['target'].value_counts()
print(class_counts)

plt.figure(figsize=(8,6))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title('Class balance check')
plt.xlabel('Class')
plt.ylabel('Count')
```

```python
# separate target from other features

features = data.drop(columns=['target'])
Target = data['target']

# data splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, Target, test_size=0.2, random_state=42)

# check the shape of the test and train parameters
print('Original Data shape', features.shape)
print('The shape of X_train is:', X_train.shape)
print('The shape of X_test is:', X_test.shape)
print('The shape of y_train is:', y_train.shape)
print('The shape of y_test is:', y_test.shape)
```

### Data Standardisation/Scaling

```python
from sklearn.preprocessing import MinMaxScaler

# Fit scaler on training data
scaler = MinMaxScaler()
# Transform training data
X_train = scaler.fit_transform(X_train)

# Transform testing data
X_test= scaler.transform(X_test)

print(X_train.shape)
print(X_test.shape)
```

### FEATURE SELECTION WITH PCA

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(13)


X_train = pca.fit_transform(X_train)

# Individual component variance
ind_var = pca.explained_variance_

# Cumulative variance
cum_var_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot individual component variance
plt.figure()
plt.bar(range(1, len(ind_var) + 1), ind_var, alpha=0.7)
plt.xlabel('Pri. Component')
plt.ylabel('Var. Explained')
plt.title('Individual Component Var.')
plt.show()

# Plot cumulative variance explained
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(cum_var_ratio) + 1), cum_var_ratio, marker='o', 
         linestyle='--', color='r')
plt.xlabel('No. of PCs')
plt.ylabel('Cumulativ Var. Explained')
plt.title('Cumulative Var. Explained by PCs')
plt.grid()
plt.show()

X_test = pca.transform(X_test)

#------------------------------------
```

### K NEAREST NEIGHBORS (KNN) ALGORITHM

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn_train_accuracy = []
knn_test_accuracy = []


settings = range(1,10)
for n in settings:
    
    #building the model
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train, y_train)
    
    #append training and generalisation accuracy
    knn_train_accuracy.append(knn.score(X_train, y_train))
    knn_test_accuracy.append(knn.score(X_test, y_test))
    
    #plotting confusion matrix for each n_neigbhour parameter
    predicted = knn.predict(X_test)
    confusion_matrix = metrics.confusion_matrix(y_test, predicted)
    # Plot
    plt.figure()  
    met = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[0, 1])
    met.plot()
    plt.title(f'Confusion Matrix {n} Neighbours')
    plt.show()

plt.plot(settings, knn_train_accuracy, label = 'Train Accuracy')
plt.plot(settings, knn_test_accuracy, label = 'Test Accuracy')
plt.ylabel('Acuracy'),
plt.xlabel('n_neighbors')
plt.grid()
plt.legend()

#print(knn_train_accuracy)
print(knn_test_accuracy)
```

### Evaluation: Precision, Specificity and Sensitivity for KNN

```python
from sklearn.metrics import classification_report

#y_pred=NB_clsifr.predict(X_test)
print(classification_report(y_test, predicted))

# specificity
TN,FP,FN,TP = (metrics.confusion_matrix(y_test, predicted)).ravel()
specificity= TN/(TN+FP)
print('knn specificity is', specificity)
```

### RANDOM FOREST ALGORITHM (RF)

```python
from sklearn.ensemble import RandomForestClassifier

train_accuracy = []
test_accuracy = []


settings = range(1,10)
for n in settings:
    
    #building the model
    RF = RandomForestClassifier(n_estimators=n, max_features='sqrt', random_state=42)
    RF.fit(X_train, y_train)
    
    #append training and generalisation accuracy
    train_accuracy.append(RF.score(X_train, y_train))
    test_accuracy.append(RF.score(X_test, y_test))
    
    #plotting confusion matrix for each n_neigbhour parameter
    predicted = RF.predict(X_test)
    confusion_matrix = metrics.confusion_matrix(y_test, predicted)
    # Plot
    plt.figure()  
    met = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[0, 1])
    met.plot()
    plt.title(f'Confusion Matrix {n}')
    plt.show()

plt.plot(settings, train_accuracy, label = 'Train Accuracy')
plt.plot(settings, test_accuracy, label = 'Test Accuracy')
plt.ylabel('Accuracy'),
plt.xlabel('n_estimators')
plt.grid()
plt.legend()

#print(train_accuracy)
print(test_accuracy)
```

### Ensemble learning using voting classifier of all four models

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

RF = RandomForestClassifier(random_state=42)
KNN = KNeighborsClassifier()
NB = GaussianNB()
SVM = SVC(kernel = 'rbf', random_state=42)

vc = (VotingClassifier(estimators=[("RF", RF),("KNN",KNN),("NB",NB),("SVM",SVM)]))

vc.fit(X_train, y_train)

y_train_pred = vc.predict(X_train)

print(accuracy_score(y_train, y_train_pred))

# Predict the test labels
y_test_pred = vc.predict(X_test)

# Calculate the test accuracy
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Test Accuracy:", test_accuracy)

y_test_pred=vc.predict(X_test)
print(classification_report(y_test,y_test_pred))
```

### Ensemble learning using voting classifier of the best two performing models RF and NB

```python
RF = RandomForestClassifier(random_state=42)
NB = GaussianNB()

vc = (VotingClassifier(estimators=[("RF", RF),("NB",NB),]))

vc.fit(X_train, y_train)

y_train_pred = vc.predict(X_train)

print(accuracy_score(y_train, y_train_pred))

# Predict the test labels
y_test_pred = vc.predict(X_test)

# Calculate the test accuracy
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Test Accuracy:", test_accuracy)

y_test_pred=vc.predict(X_test)
print(classification_report(y_test,y_test_pred))
```

### Hyper parameter tunning the best models Random Forest and Naive Bayes using random search

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report

# Random Forest
```

```python
param_grid = { 
    'n_estimators': [25, 50, 100, 150], 
    'max_features': ['sqrt', 'log2', None], 
    'max_depth': [3, 6, 9], 
    'max_leaf_nodes': [3, 6, 9], 
} 

random_search = RandomizedSearchCV(RandomForestClassifier(), 
                                   param_grid) 
random_search.fit(X_train, y_train)

# Print the best hyperparameters and best accuracy score for Random Search
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)

best_rf_random = random_search.best_estimator_
print(best_rf_random) 
```

```python
model_random = RandomForestClassifier(max_depth=6, 
                                      max_features='log2', 
                                      max_leaf_nodes=6, 
                                      n_estimators=25) 
model_random.fit(X_train, y_train) 
y_pred_rand = model_random.predict(X_test) 

test_accuracy_random = best_rf_random.score(X_test, y_test)
print("Test Accuracy (Random Search): {:.2f}%".format(test_accuracy_random * 100))
print(classification_report(y_pred_rand, y_test)) 
```
